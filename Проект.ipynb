```python
#как решить проблему гетероскедастичности
import numpy as np
import statsmodels.api as sm

# Создание матрицы признаков x и вектора целевой переменной y

# Определение весов
weights = 1.0 / np.power(y, 1)

# Построение модели с использованием взвешенных наименьших квадратов
lin_reg = sm.WLS(y, x, weights=weights)
lin_reg = lin_reg.fit()

# Вывод результатов
print(lin_reg.summary())
```
